```python
def bin2int(x):
    tot = 0
    tot -= 2**(len(x) - 1) * int(x[0])
    for i in range(len(x)):
        if i == 0:
            continue
        tot += 2**(len(x) - i - 1) * int(x[i])
    return tot
```
